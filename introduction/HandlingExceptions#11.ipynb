# Handle Exceptions in Python 

```python
try:
    file = open('bidule.txt', 'r')
except:
    print("file not exist")
    
else:
    print(file.read())
    
file.close()
```

```python
value = input("Give a number: ")
try:
    result = float(value)/2 
except:
    print("Conversion not possible")

else:
    print(result)
```

#### Exception Handling Flow

#### Some types of exception

```python
try:
    file = open('data.txt', 'r')
    list_values = file.readlines()

except FileNotFoundError as fnfe:
    print(fnfe)
    
else:
    for v in list_values:
        try:
            result = int(v)/2
            
        except ValueError as ve:
            print(ve)
        else:
            print("for {} result: {}\n".format(v, result))
finally:
    print("Closing the file")
    file.close()
```
