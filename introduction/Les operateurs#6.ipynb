# Les opérateurs en Python

## Quelques types d'opérateurs

1. Opérateurs Arithmétiques
    
2. Opérateurs de comparaison

3. Opérateurs logiques/Booléens

4. Opérateurs d'assignation/affectation

5. Opérateurs d'appartenance 

## 1. Opérateurs Arithmétiques ( + , -, *, /, %, **)   
Utilisé pour des opérations arithmétiques 

```python
a = 5; b = 2

#Addition
print(a+b)

#Soustraction(-)
print(a-b)

#Multiplication(*)
print(a*b)

#Division(/)
print(a/b)

#Division modulaire (%)
print(a%b)

#Puissance(**)
print(b**a)
```

# 2. Opérateurs de comparaison (>, <, ==, !=, >=, <=)   
Utilisés pour la comparaison des valeurs. Les résultat est **True** ou **False** selon les situations.  

```python
a = 10; b = 20; c = 2; d = 10        

#Vérifier si c est strictement supérieur à a
print(c<a)

#Vérifier si a est supérieur ou égal à d
print(a>=d)

#Vérifier si c est différent de d
print(c!=d)
```

# 3. Opérateurs logiques / booléens (not, or, and)  


```python
a = True; b = False

#print a and b
print(a and b)

#print a or b
print(a or b)

#print not b 
print(not(b))

#print not(b) and a 
print(not(b) and a)
```

# 4. Opérateurs d'affectation (    =,  +=,  -=,  *=,  /=,  %=, **=)  
Utilisé pour attribuer une valeur à une variable.  

```python
a = 3; b = 2

#Addition et affectation
a += 5  # a+5 ==> a  ; a = a+5   
print(a)

#Soustraction et affectation
b -= 2 # b = b - 2
print(b)


#Multiplication et affectation
c = 4
c *=2 # c=c*2
print(c)

#Division et affectation
d=5
d /=2 # d = d/2
print(d)

#Modulo et affectation
e=8
e %=3
print(e)

#Puissance et affectation
f = 3
f **=2 # f = f**2
print(f)
```

# 5. Opérateurs d'appartenance (in et not in )

Utilisé pour vérifier l'existence d'une valeur ou une variable dans une chaine de caractère, une liste, un tuple, un set ou un dictionnaire. 

```python
liste = [1, 2, 3, 4]

#Vérifier si 5 est présent dans la liste
print(5 in liste)

#Vérifier si 4 est présent dans la liste
print(4 in liste)
```

```python
d = {"a": 1, "b": 2}

#Vérifier si la clé "b" existe dans le dictionnaire d 
print("b" not in d)
```
